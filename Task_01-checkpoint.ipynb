```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
population_file = r"C:\Users\THADAKALURU JASWANTH\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_23\API_SP.POP.TOTL_DS2_en_csv_v2_23.csv"
metadata_country_file = r"C:\Users\THADAKALURU JASWANTH\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_23\Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_23.csv"
metadata_indicator_file = r"C:\Users\THADAKALURU JASWANTH\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_23\Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_23.csv"
# df_population = pd.read_csv(population_file, skiprows=4)
df_population = pd.read_csv(population_file, skiprows=4)
df_metadata_country = pd.read_csv(metadata_country_file)
df_metadata_indicator = pd.read_csv(metadata_indicator_file)

# Process the data
year = "2022"
df_year = df_population[['Country Name', year]].dropna()

# Due to the large number of countries, we'll plot the top 20 countries by population for clarity
df_top_20 = df_year.sort_values(by=year, ascending=False).head(20)

# Visualize the data
plt.figure(figsize=(14, 8))
plt.bar(df_top_20['Country Name'], df_top_20[year], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title(f'Population Distribution by Country in {year}')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample hypothetical dataset
data = {
    'Country': ['Country1'] * 10 + ['Country2'] * 10,
    'Age Group': ['0-4', '5-9', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40-44', '45-49'] * 2,
    'Gender': ['Male', 'Female'] * 5 + ['Male', 'Female'] * 5,
    'Population': [1000, 950, 1100, 1050, 1200, 1150, 1300, 1250, 1400, 1350, 1500, 1450, 1600, 1550, 1700, 1650, 1800, 1750, 1900, 1850]
}

df = pd.DataFrame(data)

# Group by Age Group and sum populations
age_distribution = df.groupby('Age Group')['Population'].sum().sort_index()

# Visualize the age distribution
plt.figure(figsize=(12, 6))
age_distribution.plot(kind='bar', color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Population')
plt.title('Age Distribution in Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Group by Age Group and Gender, then sum populations
gender_distribution = df.groupby(['Age Group', 'Gender'])['Population'].sum().unstack()

# Visualize the gender distribution
gender_distribution.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.xlabel('Age Group')
plt.ylabel('Population')
plt.title('Gender Distribution by Age Group')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
